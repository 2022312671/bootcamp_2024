# 3.	**Seasonal** **Trends**

•	How do the number of pickups vary across different seasons (winter, spring, summer, fall)?

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset (replace with actual data loading)
data = {
    'pickup_dt': ['1/1/2015 1:00', '4/1/2015 2:00', '7/1/2015 3:00', '10/1/2015 4:00', '1/1/2016 5:00'],
    'pickups': [152.0, 1519.0, 0.0, 5258.0, 405.0],
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Convert pickup_dt to datetime format
df['pickup_dt'] = pd.to_datetime(df['pickup_dt'], format='%m/%d/%Y %H:%M')

# Define seasons based on month ranges
def get_season(month):
    if 3 <= month <= 5:
        return 'Spring'
    elif 6 <= month <= 8:
        return 'Summer'
    elif 9 <= month <= 11:
        return 'Fall'
    else:
        return 'Winter'

# Extract season from pickup_dt
df['season'] = df['pickup_dt'].dt.month.apply(get_season)

# Group by season and sum the pickups
seasonal_pickups = df.groupby('season')['pickups'].sum()

# Plotting the number of pickups across different seasons
plt.bar(seasonal_pickups.index, seasonal_pickups.values)
plt.title('Number of Pickups Across Seasons')
plt.xlabel('Season')
plt.ylabel('Number of Pickups')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset (replace with actual data loading)
data = {
    'pickup_dt': ['1/1/2015 7:00', '1/1/2015 7:30', '1/1/2015 8:00', '1/1/2015 17:00', '1/1/2015 18:00'],
    'pickups': [152.0, 1519.0, 0.0, 5258.0, 405.0],
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Convert pickup_dt to datetime format
df['pickup_dt'] = pd.to_datetime(df['pickup_dt'], format='%m/%d/%Y %H:%M')

# Extract hour of day from pickup_dt
df['hour_of_day'] = df['pickup_dt'].dt.hour

# Filter data for rush hours (7-9 AM and 5-7 PM)
rush_hour_data = df[(df['hour_of_day'] >= 7) & (df['hour_of_day'] <= 9) | (df['hour_of_day'] >= 17) & (df['hour_of_day'] <= 19)]

# Group by hour of day and calculate total pickups
rush_hour_pickups = rush_hour_data.groupby('hour_of_day')['pickups'].sum()

# Visualize pickup trends during rush hours
plt.figure(figsize=(10, 6))
plt.plot(rush_hour_pickups.index, rush_hour_pickups.values, marker='o')
plt.title('Pickup Trends During Rush Hours')
plt.xlabel('Hour of Day')
plt.ylabel('Number of Pickups')
plt.xticks(range(7, 20))  # Set x-axis ticks to show hours
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Sample dataset (replace with actual data loading)
data = {
    'pickup_dt': ['1/1/2015 1:00', '1/1/2015 2:00', '1/1/2015 3:00', '1/1/2015 4:00', '1/1/2015 5:00'],
    'pickups': [152.0, 1519.0, 0.0, 5258.0, 405.0],
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Convert pickup_dt to datetime format
df['pickup_dt'] = pd.to_datetime(df['pickup_dt'], format='%m/%d/%Y %H:%M')

# Extract hour of day from pickup_dt
df['hour_of_day'] = df['pickup_dt'].dt.hour

# Filter data for late-night hours (12 AM - 4 AM)
late_night_data = df[(df['hour_of_day'] >= 0) & (df['hour_of_day'] <= 4)]

# Calculate average pickups during late-night hours
average_pickups_late_night = late_night_data['pickups'].mean()

print(f"Average Number of Pickups During Late-Night Hours: {average_pickups_late_night}")
```
